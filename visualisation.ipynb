```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("CAR DETAILS FROM CAR DEKHO.csv")
df.head()
```

```python
sns.histplot(np.log(df["selling_price"]),kde=True)
```

```python
print(np.log1p(df["selling_price"]).skew())
df["selling_price"]=np.log1p(df["selling_price"])
```

```python
arr=df["fuel"].value_counts()
plt.bar(arr.index,arr.values,color="orange")
print(arr)
```

```python
arr=df["owner"].value_counts()
print(arr)
```

```python
plt.pie(arr.values,labels=arr.index,autopct="%1.2f%%",explode=[0.10]*len(arr))
plt.show()
```

```python
label=LabelEncoder()
for i in df[["name","fuel","seller_type","transmission","owner"]]:
    df[i]=label.fit_transform(df[i])
```

```python
sns.heatmap(df.corr(),annot=True)
```
